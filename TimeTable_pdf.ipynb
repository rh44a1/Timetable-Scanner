```python
with pdfplumber.open(pdf_path) as pdf:
    page = pdf.pages[0]
    table = page.extract_table()

table
```

```python

raw_table = table


days = {"MON", "TUE", "WED", "THU", "FRI"}

timetable_rows = [
    row for row in raw_table
    if row and row[0] in days
]

timetable_rows
```

```python
import re

periods = ["I", "II", "III", "IV", "V", "VI", "VII", "VIII"]

period_col_map = {
    "I": 1,
    "II": 2,
    "III": 4,
    "IV": 5,
    "V": 7,
    "VI": 8,
    "VII": 10,
    "VIII": 11
}

course_pattern = r'(?:UCS|UMA|UGA|UPA|UCSV)\d{4}'

clean_timetable = {}

for row in timetable_rows:
    day = row[0]
    clean_timetable[day] = {}

    previous_subject = None

    for period in periods:
        col_idx = period_col_map[period]
        cell = row[col_idx]

        if cell:
            codes = re.findall(course_pattern, cell)
            subject = " / ".join(codes) if codes else None
        else:
            subject = None


        if subject is None and previous_subject:

            if " / " in previous_subject:
                subject = previous_subject

        clean_timetable[day][period] = subject
        previous_subject = subject

clean_timetable
```

```python
subject_info={}
for row in table:
  if not row or not row[0]:
    continue
  code=row[0]
  if not isinstance(code,str):
    continue
  if not any(code.startswith(p) for p in ["UCS","UMA","UGA","UPA","UCSV"]):
    continue
  subject_info[code]={
      "name":row[1],
      "faculty":row[4],
      "faculty_initial":row[8],
      "category": row[10],
      "credits": row[11]
  }
subject_info
```

```python
detailed_timetable = {}

for day, periods in clean_timetable.items():
    detailed_timetable[day] = {}

    for period, value in periods.items():
        if value is None:
            detailed_timetable[day][period] = None
            continue

        codes = [c.strip() for c in value.split("/")]
        details = []

        for code in codes:
            if code in subject_info:
                info = subject_info[code]
                details.append({
                    "code": code,
                    "name": info["name"],
                    "faculty": info["faculty"],
                    "faculty_initial": info["faculty_initial"]
                })
            else:
                details.append({
                    "code": code,
                    "name": None,
                    "faculty": None,
                    "faculty_initial": None
                })


        detailed_timetable[day][period] = details


detailed_timetable
```

```python
import json

with open("timetable.json", "w") as f:
    json.dump(detailed_timetable, f, indent=2)
```
